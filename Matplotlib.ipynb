question 1

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(0,10,10)
y=np.random.randint(0,10,10)
plt.scatter(x,y)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
```

question 2

A scatter plot is a type of graph that shows the relationship between two variables in a data set

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
```

question 3

Subplots in matplotlib allow you to plot multiple plots in a single figure.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,4,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,4,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(1,4,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,4)
plt.plot(x,y)
```

question 4

A bar plot or bar chart is a graph that represents the category of data with rectangular bars

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
